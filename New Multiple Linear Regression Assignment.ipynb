```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm

import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("Social Media Dummy Data HSS.csv")
data = raw_data.copy()
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
x1 = data[["TV PROMOTION BUDGET in millions","Radio","Social Media"]]
y = data["Sales"]
```

```python
x = sm.add_constant(x1)
x.head()
```

```python
model = sm.OLS(y,x).fit()
model.summary()
```

```python
y_cap = model.params[0]  + (model.params[1] * 16.0) + (model.params[2] * 6.566231) + (model.params[3] * 2.907983)
y_cap
```
