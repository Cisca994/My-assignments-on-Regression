```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

sns.set()
```

```python
raw_data = pd.read_csv("Salary_dataset.csv")
data = raw_data.copy()
```

```python
x1 = data["YearsExperience"]
y = data["Salary"]
```

```python
import statsmodels.api as sm
```

```python
plt.scatter(x1, y)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Salary chart")
plt.show()
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y, x).fit()
```

```python
result.summary()
```

```python
result.params[1]
```

```python
result.predict()
```

```python
y_hat = result.params[0] + result.params[1] * x1
y_hat
```

```python
data["Predictions"] = y_hat
```

```python
data["Error"] = data["Salary"] - data["Predictions"]
data
```
